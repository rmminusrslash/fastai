## Visualize data

```python
# DataBlock: Generic container to quickly build `Datasets` and `DataLoaders`

mapping={
    "women fashion Seventies Look":0,
    "women fashion draping":1,
    "women fashion Bodysuits":2
}
def get_y(foldername):
    '''
    we need to return a list of strings to use the target as a multi label target
    (even when we only have one target)

    '''
    foldername=parent_label(foldername)
    return [foldername]


fashion = DataBlock(
    blocks=(ImageBlock, MultiCategoryBlock),  # changed CategoryBlock to MultiCategoryBlock
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=get_y, # changed parent_label to get_y
    item_tfms=Resize(128))
```

```python
# print to check the output
# indeed we managed to get a one hot encoded target with three values
ds=fashion.datasets(path)
```

```python
dls = fashion.dataloaders(path)
```

```python
import matplotlib
font = {'size'   : 8}
matplotlib.rc('font', **font)
dls.valid.show_batch(max_n=20, nrows=3)
```

```python
# add and visualize image augmentation
fashion = fashion.new(item_tfms=Resize(128), batch_tfms=aug_transforms())
dls = fashion.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

```python
# also add random cropping on the original image
# resize already crops. it is better to add a bit more variety than using the same crop
fashion = fashion.new(
    item_tfms=RandomResizedCrop(224,min_scale=0.5),
    batch_tfms=aug_transforms())
dls = fashion.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

## Learn model

```python
def multi_acc(preds,targets,threshold=0.5):
    return ((preds>threshold)==targets).float().mean()
```

```python
learn = cnn_learner(dls, resnet18, metrics=partial(multi_acc, threshold=0.2))
# Fine tune with `freeze` for `freeze_epochs` then with `unfreeze` from `epochs` using discriminative LR
learn.fine_tune(5)
```

```python
preds,targets=learn.get_preds()

xs=np.linspace(0.05,0.95,29)
acc=[multi_acc(preds,targets,threshold=i) for i in xs]
plt.plot(xs,acc)
```
