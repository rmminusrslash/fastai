## Gather data

```python
from fastbook import *
from fastai.vision.widgets import *
```

```python
# trends 2020
fashion_types = 'Seventies Look','draping', 'Bodysuits'
path = Path('fashion')
```

```python
fashion_types
```

```python
from bing_image_downloader import downloader
if not path.exists():
    path.mkdir()
for o in fashion_types:
    query="women fashion "+ o
    print("query",query)
    downloader.download(query, limit=150,  output_dir="fashion", adult_filter_off=False, 
                force_replace=True, timeout=60)
```

### Check that files can be opened 
Basic data integrity check

```python
fns = get_image_files(path)
fns
```

```python
# check if they can all be opened
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink);
```

## Visualize data

```python
# DataBlock: Generic container to quickly build `Datasets` and `DataLoaders`
fashion = DataBlock(
    blocks=(ImageBlock, CategoryBlock), 
    get_items=get_image_files, 
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128))
```

```python
dls = fashion.dataloaders(path)
```

```python
import matplotlib
font = {'size'   : 8}
matplotlib.rc('font', **font)
dls.valid.show_batch(max_n=20, nrows=3)
```

```python
# add and visualize image augmentation
fashion = fashion.new(item_tfms=Resize(128), batch_tfms=aug_transforms())
dls = fashion.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

```python
# also add random cropping on the original image
# resize already crops. it is better to add a bit more variety than using the same crop
fashion = fashion.new(
    item_tfms=RandomResizedCrop(224,min_scale=0.5),
    batch_tfms=aug_transforms())
dls = fashion.dataloaders(path)
dls.train.show_batch(max_n=8, nrows=2, unique=True)
```

## Learn model

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
# Fine tune with `freeze` for `freeze_epochs` then with `unfreeze` from `epochs` using discriminative LR
learn.fine_tune(5)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```
